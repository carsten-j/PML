````python
### 1. Import Necessary Libraries
import matplotlib.pyplot as plt
import numpy as np
import torch
import torch.nn as nn
import torch.optim as optim
import torchvision.datasets as datasets
import torchvision.transforms as transforms
from torch.utils.data import DataLoader
````

````python
### 2. Define Hyperparameters
batch_size = 128
learning_rate = 1e-3
num_epochs = 20
latent_dim = 128
````

````python
### 3. Prepare the CIFAR-10 Dataset

# Data transformations
transform = transforms.Compose(
    [transforms.ToTensor(), transforms.Normalize((0.5,), (0.5,))]
)

# Load the CIFAR-10 dataset
train_dataset = datasets.CIFAR10(
    root="./data", train=True, transform=transform, download=True
)
test_dataset = datasets.CIFAR10(
    root="./data", train=False, transform=transform, download=True
)

# Data loaders
train_loader = DataLoader(dataset=train_dataset, batch_size=batch_size, shuffle=True)
test_loader = DataLoader(dataset=test_dataset, batch_size=batch_size, shuffle=False)
````

````python
### 4. Define the Autoencoder Model


class ConvAutoencoder(nn.Module):
    def __init__(self, latent_dim):
        super(ConvAutoencoder, self).__init__()
        # Encoder
        self.encoder = nn.Sequential(
            nn.Conv2d(3, 32, kernel_size=4, stride=2, padding=1),  # [B, 32, 16, 16]
            nn.BatchNorm2d(32),
            nn.LeakyReLU(),
            nn.Conv2d(32, 64, kernel_size=4, stride=2, padding=1),  # [B, 64, 8, 8]
            nn.BatchNorm2d(64),
            nn.LeakyReLU(),
            nn.Conv2d(64, 128, kernel_size=4, stride=2, padding=1),  # [B, 128, 4, 4]
            nn.BatchNorm2d(128),
            nn.LeakyReLU(),
            nn.Flatten(),  # [B, 128*4*4]
            nn.Linear(128 * 4 * 4, latent_dim),
        )

        # Decoder
        self.decoder = nn.Sequential(
            nn.Linear(latent_dim, 128 * 4 * 4),
            nn.Unflatten(1, (128, 4, 4)),  # [B, 128, 4, 4]
            nn.ConvTranspose2d(
                128, 64, kernel_size=4, stride=2, padding=1
            ),  # [B, 64, 8, 8]
            nn.BatchNorm2d(64),
            nn.LeakyReLU(),
            nn.ConvTranspose2d(
                64, 32, kernel_size=4, stride=2, padding=1
            ),  # [B, 32, 16, 16]
            nn.BatchNorm2d(32),
            nn.LeakyReLU(),
            nn.ConvTranspose2d(
                32, 3, kernel_size=4, stride=2, padding=1
            ),  # [B, 3, 32, 32]
            nn.Tanh(),
        )

    def forward(self, x):
        # Encode
        latent = self.encoder(x)
        # Decode
        x_reconstructed = self.decoder(latent)
        return x_reconstructed
````

````python
### 5. Define the Loss Function

# We use the Mean Squared Error (MSE) loss function to measure the reconstruction error between the input images and the reconstructed images.

criterion = nn.MSELoss()
````

````python
### 6. Initialize the Model, Optimizer, and Device

# Check for CUDA
if torch.cuda.is_available():
    device = torch.device("cuda")
# Check for MPS (for Apple Silicon)
elif torch.backends.mps.is_available():
    device = torch.device("mps")
else:
    device = torch.device("cpu")

print(f"Using device: {device}")
````

````python
model = ConvAutoencoder(latent_dim).to(device)
optimizer = optim.Adam(model.parameters(), lr=learning_rate)
````

````python
### 7. Training Loop

model.train()
train_losses = []
for epoch in range(1, num_epochs + 1):
    total_loss = 0
    for batch_idx, (data, _) in enumerate(train_loader):
        data = data.to(device)
        optimizer.zero_grad()
        reconstructed = model(data)
        loss = criterion(reconstructed, data)
        loss.backward()
        total_loss += loss.item() * data.size(0)
        optimizer.step()
    avg_loss = total_loss / len(train_loader.dataset)
    train_losses.append(avg_loss)
    print(f"Epoch [{epoch}/{num_epochs}], Loss: {avg_loss:.4f}")
````

````python
torch.save(model.state_dict(), "models/model.pth")
````

````python
### 8. Visualizing the Results

#### Function to Display Original and Reconstructed Images


def show_reconstructions(model, data_loader):
    model.eval()
    with torch.no_grad():
        data_iter = iter(data_loader)
        images, _ = next(iter(data_iter))
        images = images.to(device)
        reconstructed = model(images)
        # Move images to CPU and convert to numpy
        images = images.cpu().numpy()
        reconstructed = reconstructed.cpu().numpy()

        # Display original and reconstructed images
        fig, axes = plt.subplots(2, 8, figsize=(15, 4))
        for i in range(8):
            # Original images
            ax = axes[0, i]
            img = images[i].transpose(1, 2, 0) * 0.5 + 0.5  # Unnormalize
            ax.imshow(img)
            ax.axis("off")
            ax.set_title("Original")

            # Reconstructed images
            ax = axes[1, i]
            img = reconstructed[i].transpose(1, 2, 0) * 0.5 + 0.5  # Unnormalize
            img = np.clip(img, 0, 1)
            ax.imshow(img)
            ax.axis("off")
            ax.set_title("Reconstructed")
        plt.show()
````

````python
#### Display the Reconstructions

show_reconstructions(model, test_loader)
````

### **Code Responsible for Normalizing Input Images:**

```python
# Data transformations
transform = transforms.Compose([
    transforms.ToTensor(),
    transforms.Normalize((0.5,), (0.5,))
])
```

This code is part of the data preprocessing steps applied to the CIFAR-10 dataset before it's fed into the autoencoder model.

---

### **Explanation of the Code:**

1. **`transforms.ToTensor()`**

   - **Purpose:** Converts a PIL Image or NumPy `ndarray` into a PyTorch Tensor.
   - **Effect on Pixel Values:** Scales pixel values from the range **$$0, 255$$** (for images loaded as uint8) to **$$0.0, 1.0$$** (as floating-point numbers).
   - **Process:**
     - Each pixel value is divided by 255.
     - For example, a pixel value of 128 would become `128 / 255 ≈ 0.50196`.

2. **`transforms.Normalize((0.5,), (0.5,))`**

   - **Purpose:** Normalizes the tensor image with mean and standard deviation.
   - **Parameters:**
     - **Mean:** `(0.5,)`
     - **Standard Deviation:** `(0.5,)`
   - **Effect on Pixel Values:** Transforms pixel values from **$$0.0, 1.0$$** to **$$-1.0, 1.0$$**.
   - **Process:**
     - The normalization is applied using the formula:
       $$
       \text{Normalized Value} = \frac{(\text{Original Value} - \text{Mean})}{\text{Standard Deviation}}
       $$
     - Since both mean and standard deviation are `0.5`, the formula simplifies to:
       $$
       \text{Normalized Value} = ( \text{Original Value} - 0.5 ) \times 2
       $$
     - This effectively maps the range from $$0.0, 1.0$$ to $$-1.0, 1.0$$.

---

### **Detailed Step-by-Step Calculation:**

Let's walk through the transformation for a sample pixel value.

**Assume a pixel value in the original image:**

- **Original pixel value in uint8 format:** `x_uint8 = 128`
- **After `ToTensor()` conversion:**
  - $$
  x_{\text{tensor}} = \frac{x_{\text{uint8}}}{255} = \frac{128}{255} \approx 0.50196
  $$

**Applying `Normalize((0.5,), (0.5,))`:**

- **Normalization formula:**
  - $$
  x_{\text{normalized}} = \frac{(x_{\text{tensor}} - 0.5)}{0.5}
  $$
- **Calculating normalized value:**
  - $$
  x_{\text{normalized}} = \frac{(0.50196 - 0.5)}{0.5} \approx \frac{0.00196}{0.5} \approx 0.00392
  $$
- **Simplified calculation (since std = 0.5):**
  - $$
  x_{\text{normalized}} = (x_{\text{tensor}} - 0.5) \times 2
  $$
  - $$
  x_{\text{normalized}} = (0.50196 - 0.5) \times 2 \approx 0.00392
  $$
  
**Resulting Pixel Value:**

- The pixel value after normalization is approximately `0.00392`, falling within the range $$-1.0, 1.0$$.

**Edge Cases:**

- **Minimum Pixel Value (0 in uint8):**
  - After `ToTensor()`: `0 / 255 = 0.0`
  - After normalization: `(0.0 - 0.5) / 0.5 = -1.0`
- **Maximum Pixel Value (255 in uint8):**
  - After `ToTensor()`: `255 / 255 = 1.0`
  - After normalization: `(1.0 - 0.5) / 0.5 = 1.0`

---

### **Why Normalize Between -1 and 1?**

- **Matching Activation Functions:**
  - The decoder's final layer uses the `Tanh` activation function, which outputs values in the range $$-1, 1$$.
  - Normalizing input images to the same range ensures consistency between the input and output distributions.
  
- **Training Stability:**
  - Normalizing data often leads to faster convergence and better training stability in neural networks.
  - It helps in preventing issues like vanishing or exploding gradients.

- **Standard Practice:**
  - For models using activation functions with specific output ranges (like `Tanh`), it's standard practice to normalize inputs to match those ranges.

Certainly! Let's delve into the details of how the dimensions of the input image change after applying the convolutional layer in your neural network. We'll start by understanding the initial dimensions and then walk through the computation step by step to arrive at the new dimensions.

---

### **Initial Image Dimensions:**

- **CIFAR-10 Image Size:** Each image in the CIFAR-10 dataset is **32 pixels high** and **32 pixels wide** with **3 color channels (Red, Green, Blue)**.
- **Tensor Representation in PyTorch:** Images are represented as tensors with dimensions `[Channels, Height, Width]`, so each image has the dimensions:
  $$
  [C, H, W] = [3, 32, 32]
  $$

- **Batch Size:** When processing images in batches, an additional dimension is added for the batch. If the batch size is **`B`**, the images are represented as:
  $$
  [B, C, H, W] = [B, 3, 32, 32]
  $$

---

### **Convolutional Layer Specifications:**

The convolutional layer in question is defined as:

```python
nn.Conv2d(in_channels=3, out_channels=32, kernel_size=4, stride=2, padding=1)
```

- **Input Channels (`in_channels`):** 3 (matching the 3 color channels of the image)
- **Output Channels (`out_channels`):** 32 (number of feature maps produced by the convolution)
- **Kernel Size (`kernel_size`):** 4 (the size of the convolutional filter is 4x4)
- **Stride (`stride`):** 2 (the filter moves 2 pixels at a time)
- **Padding (`padding`):** 1 (padding of 1 pixel is added to each side of the input)

---

### **Calculating the Output Dimensions:**

The output dimensions after a convolutional layer can be calculated using the following formula for each spatial dimension (height and width):

$$
\text{Output Size} = \left\lfloor \frac{\text{Input Size} + 2 \times \text{Padding} - \text{Dilation} \times (\text{Kernel Size} - 1) - 1}{\text{Stride}} + 1 \right\rfloor
$$

- **Note:**
  - The default dilation is 1 unless specified otherwise.
  - The floor function $\lfloor \cdot \rfloor$ indicates that the result is rounded down to the nearest integer.

#### **Parameters:**

- **Input Size (`H_in`, `W_in`):** 32 (since both height and width are the same)
- **Kernel Size (`K`):** 4
- **Stride (`S`):** 2
- **Padding (`P`):** 1
- **Dilation (`D`):** 1 (default value)

#### **Compute the Output Height and Width:**

Let's compute the output size step by step.

1. **Plug in the Values:**

   $$
   \text{Output Size} = \left\lfloor \frac{32 + 2 \times 1 - 1 \times (4 - 1) - 1}{2} + 1 \right\rfloor
   $$

2. **Simplify the Expression:**

   - Compute the terms inside the numerator:

     - $2 \times \text{Padding} = 2 \times 1 = 2$
     - $\text{Dilation} \times (\text{Kernel Size} - 1) = 1 \times (4 - 1) = 3$
     - $\text{Input Size} + 2 \times \text{Padding} = 32 + 2 = 34$

   - Now, the numerator becomes:

     $$
     34 - 3 - 1 = 30
     $$

3. **Divide by the Stride and Add 1:**

   $$
   \frac{30}{2} + 1 = 15 + 1 = 16
   $$

4. **Apply the Floor Function:**

   - Since the result is already an integer, the floor function doesn't change it.

   $$
   \text{Output Size} = \left\lfloor 16 \right\rfloor = 16
   $$

#### **Conclusion:**

- **Output Height (`H_out`):** 16
- **Output Width (`W_out`):** 16

- **Number of Output Channels:** 32 (as specified by `out_channels`)

- **Output Tensor Dimensions:**

  $$
  [B, \text{Output Channels}, H_{\text{out}}, W_{\text{out}}] = [B, 32, 16, 16]
  $$

---

### **Complete Calculation for Height and Width:**

For clarity, here is the calculation for both height and width step by step.

#### **Given:**

- Input Size: $ H_{\text{in}} = W_{\text{in}} = 32 $
- Padding: $ P = 1 $
- Kernel Size: $ K = 4 $
- Dilation: $ D = 1 $
- Stride: $ S = 2 $

#### **Formula:**

$$
\text{Output Size} = \left\lfloor \frac{H_{\text{in}} + 2P - D(K - 1) - 1}{S} + 1 \right\rfloor
$$

#### **Calculations:**

1. **Compute Numerator:**

   $$
   H_{\text{in}} + 2P - D(K - 1) - 1 = 32 + 2 \times 1 - 1 \times (4 - 1) - 1 = 32 + 2 - 3 - 1 = 30
   $$

2. **Divide by Stride and Add 1:**

   $$
   \frac{30}{2} + 1 = 15 + 1 = 16
   $$

3. **Apply Floor Function:**

   $$
   \text{Output Size} = \left\lfloor 16 \right\rfloor = 16
   $$

---

### **Understanding the Convolution Operation:**

To further understand why these calculations yield the output dimensions, let's consider the effect of each parameter:

#### **1. Padding (`P`):**

- Padding adds a border of pixels around the input image.
- With `padding=1`, a border of 1 pixel is added to all sides.
- The input size increases from $32 \times 32$ to $ (32 + 2 \times 1) \times (32 + 2 \times 1) = 34 \times 34 $.

#### **2. Kernel Size (`K`):**

- The convolution operation uses a kernel (filter) of size $4 \times 4$.
- The receptive field (region of the input that influences a single output value) is $4 \times 4$.

#### **3. Dilation (`D`):**

- With `dilation=1`, there is no spacing between kernel elements, so the kernel size remains $4 \times 4$.

#### **4. Stride (`S`):**

- The stride determines how the filter moves across the input.
- With `stride=2`, the filter moves 2 pixels at a time horizontally and vertically.
- This effectively reduces the spatial dimensions of the output compared to the input.

---

### **Effect of Convolution on Dimensions:**

1. **Horizontal and Vertical Movement:**

   - Starting from the padded input size of 34, the kernel begins at position 0.
   - The kernel moves in steps of 2 pixels due to the stride.

2. **Number of Steps:**

   - The number of steps (output size) can be thought of as:
     $$
     \text{Number of Steps} = \left\lfloor \frac{\text{Adjusted Input Size} - \text{Kernel Size}}{\text{Stride}} + 1 \right\rfloor
     $$
   - Adjusted Input Size is the input size after accounting for padding and dilation.

3. **Example Calculation:**

   - Adjusted Input Size: $34$ (after padding)
   - Kernel Size: $4$
   - Stride: $2$
   - Number of Steps:
     $$
     \left\lfloor \frac{34 - 4}{2} + 1 \right\rfloor = \left\lfloor \frac{30}{2} + 1 \right\rfloor = 16
     $$

  - **Effect on Dimensions:** **Does not change the dimensions** of the input tensor.

### **General Formula for Convolutional Layers:**

For any convolutional layer with specified parameters, you can use the following formula to compute the output dimensions:

$$
\begin{align*}
H_{\text{out}} & = \left\lfloor \frac{H_{\text{in}} + 2P_{\text{height}} - D_{\text{height}} \times (K_{\text{height}} - 1) - 1}{S_{\text{height}}} + 1 \right\rfloor \\
W_{\text{out}} & = \left\lfloor \frac{W_{\text{in}} + 2P_{\text{width}} - D_{\text{width}} \times (K_{\text{width}} - 1) - 1}{S_{\text{width}}} + 1 \right\rfloor
\end{align*}
$$

- $H_{\text{in}}$, $W_{\text{in}}$: Input height and width
- $P$: Padding
- $D$: Dilation
- $K$: Kernel size
- $S$: Stride
- $H_{\text{out}}$, $W_{\text{out}}$: Output height and width
