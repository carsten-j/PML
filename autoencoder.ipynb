# Autoencoders

An **autoencoder** is a type of artificial neural network designed to learn efficient, compressed representations of input data, typically for purposes such as dimensionality reduction, feature learning, or data denoising. The fundamental idea is to train a network to reconstruct its input, essentially learning to capture the most salient features of the data in a compressed form.



## **Architecture of Autoencoders**

An autoencoder consists of two main components:

1. **Encoder**: This part of the network compresses the input data into a lower-dimensional representation, known as the **latent space** or **bottleneck layer**. The encoder function can be represented as $h = f(x)$, where:
   - $x$ is the input data.
   - $h$ is the encoded representation.
   - $f$ is the encoding function, typically consisting of one or more layers of neurons with activation functions.

2. **Decoder**: The decoder attempts to reconstruct the original input data from the compressed representation. It takes the latent space representation $h$ and maps it back to the reconstructed input $\hat{x}$ . The decoder function can be represented as $\hat{x} = g(h)$, where:
   - $\hat{x}$ is the reconstructed input.
   - $g$ is the decoding function, typically mirroring the encoder's architecture.

The **training objective** of an autoencoder is to minimize the difference between the input $x $ and the reconstruction $\hat{x}$ . This is usually achieved by minimizing a loss function such as the mean squared error (MSE) or binary cross-entropy, depending on the nature of the data.

**Illustration of Basic Autoencoder Architecture:**

```
Input (x) → [Encoder Network] → Latent Space (h) → [Decoder Network] → Reconstruction (x̂)
```


## **Why Autoencoders are Useful**

1. **Dimensionality Reduction**:
   - **Non-linear Compression**: Unlike linear dimensionality reduction techniques like Principal Component Analysis (PCA), autoencoders can capture complex, non-linear relationships in the data due to their neural network structure.
   - **Feature Learning**: By compressing data into a lower-dimensional space, autoencoders learn meaningful features that represent the data effectively.

2. **Data Denoising**:
   - **Denoising Autoencoders**: By training autoencoders to reconstruct clean input from noisy data, they can learn to remove noise, improving data quality.

3. **Anomaly Detection**:
   - **Reconstruction Error**: Autoencoders learn to reconstruct data similar to the training set. When presented with anomaly data, reconstruction error increases, making autoencoders useful for detecting outliers.

4. **Generative Modeling**:
   - **Variational Autoencoders**: These extend autoencoders to probabilistic models, enabling the generation of new data samples similar to the training data.

5. **Pretraining and Transfer Learning**:
   - **Layer-wise Pretraining**: Autoencoders can pretrain layers of deep networks in an unsupervised manner, initializing weights that capture important features before fine-tuning on supervised tasks.

6. **Data Compression**:
   - **Efficient Storage and Transmission**: Compressed representations can reduce storage requirements and facilitate faster data transmission.


## **Types of Autoencoders and Advanced Architectures**

1. **Undercomplete Autoencoder**:
   - **Definition**: Latent space has a lower dimensionality than the input space.
   - **Purpose**: Forces the autoencoder to capture the most critical features.

2. **Overcomplete Autoencoder**:
   - **Definition**: Latent space dimension is equal to or greater than the input dimension.
   - **Regularization Needed**: Without constraints, the autoencoder might learn an identity function.

3. **Regularized Autoencoders**:
   - **Sparse Autoencoders**:
     - **Concept**: Introduce a sparsity constraint on the activation of neurons, encouraging the network to use a minimal number of active neurons to represent the data.
   - **Denoising Autoencoders**:
     - **Concept**: Train the network to reconstruct input data from a corrupted version, improving robustness to noise.
   - **Contractive Autoencoders**:
     - **Concept**: Add a penalty on the derivative of the latent space with respect to the input, encouraging the representation to be insensitive to small variations in the input.

4. **Variational Autoencoders (VAEs)**:
   - **Concept**: Model the latent space as a probability distribution (typically Gaussian). The encoder outputs parameters of the distribution (mean and variance), and the decoder reconstructs data from samples drawn from this distribution.
   - **Benefits**: Useful for generative modeling, allowing sampling of new data points.

5. **Convolutional Autoencoders**:
   - **Concept**: Use convolutional layers within the encoder and decoder, making them well-suited for image and spatial data by preserving spatial hierarchies.

6. **Recurrent Autoencoders**:
   - **Concept**: Incorporate recurrent neural networks (RNNs) for sequential data, such as time series or text.

7. **Stacked and Deep Autoencoders**:
   - **Concept**: Build deeper networks by stacking multiple layers, enabling the learning of more abstract features.


## **Common Use Cases of Autoencoders**

1. **Dimensionality Reduction**:
   - **Data Visualization**: Reducing high-dimensional data to 2D or 3D for visualization using techniques like t-SNE on the encoded representations.
   - **Computational Efficiency**: Reducing data size for faster processing in downstream tasks.

2. **Data Denoising**:
   - **Image Processing**: Removing noise from images while preserving important features.
   - **Signal Restoration**: Enhancing signals in domains like audio processing and biomedical signal analysis.

3. **Anomaly and Outlier Detection**:
   - **Network Security**: Identifying unusual patterns that may indicate security threats.
   - **Fraud Detection**: Spotting fraudulent transactions by detecting deviations from normal patterns.

4. **Generative Modeling and Data Augmentation**:
   - **Image Synthesis**: Generating new images for training data enrichment.
   - **Variational Autoencoders**: Creating new data samples in areas like drug discovery, where generating new molecular structures is valuable.

5. **Feature Extraction and Preprocessing**:
   - **Supervised Learning Preprocessing**: Using the encoder part of an autoencoder to extract features that improve classification or regression models.
   - **Transfer Learning**: Applying learned features to new but related tasks.

6. **Recommender Systems**:
   - **Collaborative Filtering**: Learning latent representations of users and items to predict preferences.

7. **Segmentation and Object Detection**:
   - **Semantic Segmentation**: Autoencoders with skip connections (e.g., U-Net architecture) can perform image segmentation tasks.

8. **Robotics and Control Systems**:
   - **State Representation**: Learning compact representations of the environment for efficient decision-making.

9. **Natural Language Processing (NLP)**:
   - **Text Reconstruction**: Learning representations of text sequences.
   - **Machine Translation**: Encoder-decoder frameworks are foundational in sequence-to-sequence models.

10. **Biomedical Applications**:
    - **Medical Imaging**: Enhancing and reconstructing images from MRI, CT scans, etc.
    - **Disease Detection**: Identifying anomalies in medical data indicative of diseases.

### **Detailed Example: Denoising Autoencoder**

**Objective**: Learn to reconstruct clean input data from its noisy counterpart.

**Process**:

1. **Corrupt the Input**: Add noise to the original data, creating a noisy input $x_{\text{noisy}}$.
2. **Encode**: Compress the noisy input through the encoder to obtain the latent representation $h = f(x_{\text{noisy}})$.
3. **Decode**: Reconstruct the clean input from the latent representation $\hat{x} = g(h)$.
4. **Train**: Optimize the network parameters to minimize the reconstruction loss between $\hat{x}$ and the original clean input $x$.

**Benefits**:

- **Robust Feature Learning**: The network learns features that are resilient to noise.
- **Improved Generalization**: By preventing the network from simply memorizing the input, it focuses on capturing underlying patterns.

## **Key Mathematical Concepts**

- **Loss Function**: Measures the discrepancy between input $x$ and reconstruction $\hat{x}$. Common choices include:
  - **Mean Squared Error (MSE)**: $\mathcal{L}(x, \hat{x}) = \frac{1}{n} \sum_{i=1}^{n} (x_i - \hat{x}_i)^2$
  - **Binary Cross-Entropy**: Used when inputs are binary or normalized between 0 and 1.

- **Regularization Terms**: Added to the loss function to impose constraints, e.g.:
  - **Sparsity Penalty**: Encourages the activation of only a subset of neurons.
  - **Contractive Penalty**: Penalizes the sensitivity of the latent representation to input variations.


## **Advantages over Traditional Methods**

- **Non-Linearity**: Ability to model non-linear relationships, unlike PCA.
- **Customizable Architectures**: Flexibility to design architectures tailored to specific data types (e.g., images, text, time series).
- **Unsupervised Learning**: Does not require labeled data, making it suitable for tasks where labeling is expensive or impractical.

## **Considerations and Limitations**

- **Risk of Learning Identity Function**: Without constraints, autoencoders may simply copy inputs to outputs without learning meaningful features.
- **Overfitting**: Autoencoders can overfit the training data, especially if the latent space is not constrained properly.
- **Choice of Hyperparameters**: Architecture design, such as the size of the latent space and regularization parameters, significantly affects performance.

## In Summary
Autoencoders are versatile neural networks that learn compressed representations of data by training to reconstruct the input. Their ability to capture essential features makes them valuable for various applications, including dimensionality reduction, feature extraction, denoising, and anomaly detection. By understanding and leveraging different types of autoencoders and architectural designs, practitioners can address complex challenges across multiple domains.

````python
# Import necessary libraries
import matplotlib.pyplot as plt  # For plotting images and graphs
import numpy as np  # For numerical operations
import torch  # PyTorch main library
import torch.optim as optim  # For optimization algorithms
import torchinfo  # For displaying model summaries
import torchvision  # For loading standard datasets
import torchvision.transforms as transforms  # For transforming data
from sklearn.model_selection import KFold  # For cross-validation
from torch import nn  # For building the neural network
from torch.utils.data import (
    DataLoader,  # For loading data in batches
    SubsetRandomSampler,  # For sampling elements randomly from a sequence
)
from torch.utils.tensorboard import SummaryWriter  # For tensorboard support
````

````python
# Check for GPU or MPS availability
device = torch.device(
    "cuda"
    if torch.cuda.is_available()
    else ("mps" if torch.backends.mps.is_available() else "cpu")
)
print("Using device:", device)
````

````python
# see https://pytorch.org/docs/stable/notes/randomness.html
# Set a random seed for reproducibility
torch.manual_seed(seed=17)
````

````python
# Define transformations for the MNIST dataset
transform = transforms.Compose(
    [
        transforms.ToTensor(),  # Converts PIL images or numpy arrays to PyTorch tensors and scales pixel values to [0, 1].
    ]
)
````

````python
# Load the MNIST dataset
train_dataset = torchvision.datasets.MNIST(
    root="./data",  # Directory to store the data
    train=True,  # Load training set
    transform=transform,  # Apply transformations
    download=True,  # Download if dataset is not present
)

test_dataset = torchvision.datasets.MNIST(
    root="./data", train=False, download=True, transform=transform
)
````

````python
# Hyperparameters
num_epochs = 10
batch_size = 64
learning_rate = 1e-3
````

````python
# Cross-validation setup
k_folds = 5
kfold = KFold(n_splits=k_folds, shuffle=True)
````

````python
class Autoencoder(nn.Module):
    def __init__(self):
        super(Autoencoder, self).__init__()
        # Encoder Layers
        self.encoder = nn.Sequential(
            nn.Conv2d(
                1, 16, kernel_size=3, stride=2, padding=1
            ),  # -> (batch, 16, 14, 14)
            nn.ReLU(True),
            nn.Conv2d(16, 4, kernel_size=3, stride=2, padding=1),  # -> (batch, 4, 7, 7)
            nn.ReLU(True),
        )
        # Decoder Layers
        self.decoder = nn.Sequential(
            nn.ConvTranspose2d(
                4, 16, kernel_size=3, stride=2, padding=1, output_padding=1
            ),  # -> (batch, 16, 14, 14)
            nn.ReLU(True),
            nn.ConvTranspose2d(
                16, 1, kernel_size=3, stride=2, padding=1, output_padding=1
            ),  # -> (batch, 1, 28, 28)
            nn.Sigmoid(),  # Compress values between 0 and 1
        )

    def forward(self, x):
        x = self.encoder(x)
        x = self.decoder(x)
        return x
````

````python
# Initialize the autoencoder and move it to the appropriate device
model = Autoencoder().to(device)
````

````python
torchinfo.summary(model, input_size=(batch_size, 1, 28, 28), device=device)
````

````python
# Define custom functions with type annotations
def train_epoch(
    model: nn.Module,
    dataloader: DataLoader,
    criterion: nn.Module,
    optimizer: optim.Optimizer,
    device: torch.device,
) -> float:
    model.train()
    running_loss = 0.0
    for data in dataloader:
        inputs, _ = data
        inputs = inputs.to(device)
        optimizer.zero_grad()

        outputs = model(inputs)
        loss = criterion(outputs, inputs)

        loss.backward()
        optimizer.step()

        running_loss += loss.item() * inputs.size(0)

    epoch_loss = running_loss / len(dataloader.dataset)
    return epoch_loss
````

````python
def validate_epoch(
    model: nn.Module, dataloader: DataLoader, criterion: nn.Module, device: torch.device
) -> float:
    model.eval()
    running_loss = 0.0
    with torch.no_grad():
        for data in dataloader:
            inputs, _ = data
            inputs = inputs.to(device)

            outputs = model(inputs)
            loss = criterion(outputs, inputs)

            running_loss += loss.item() * inputs.size(0)

    epoch_loss = running_loss / len(dataloader.dataset)
    return epoch_loss
````

````python
# Training and validation with k-fold cross-validation
criterion = nn.MSELoss()
optimizer = torch.optim.Adam(
    model.parameters(), lr=1e-3, weight_decay=1e-5
)  # Optimizer
````

````python
# Initialize TensorBoard writer
writer = SummaryWriter("runs/autoencoder_experiment")

# Begin k-fold cross-validation
for fold, (train_ids, val_ids) in enumerate(kfold.split(train_dataset)):
    print(f"Fold {fold + 1}/{k_folds}")

    # Sample elements randomly from a given list of ids
    train_subsampler = SubsetRandomSampler(train_ids)
    val_subsampler = SubsetRandomSampler(val_ids)

    # Define data loaders for training and validation in this fold
    train_loader = DataLoader(
        train_dataset, batch_size=batch_size, sampler=train_subsampler
    )
    val_loader = DataLoader(
        train_dataset, batch_size=batch_size, sampler=val_subsampler
    )

    # Training loop for this fold
    for epoch in range(num_epochs):
        train_loss = train_epoch(model, train_loader, criterion, optimizer, device)
        val_loss = validate_epoch(model, val_loader, criterion, device)

        # Log the losses to TensorBoard
        writer.add_scalars(
            f"Fold_{fold + 1}/Losses",
            {
                "Training": train_loss,
                "Validation": val_loss,
            },
            epoch,
        )

        print(
            f"Epoch {epoch + 1}/{num_epochs}: Training Loss={train_loss:.6f}, Validation Loss={val_loss:.6f}"
        )

    # Optionally, save the model for each fold
    torch.save(model.state_dict(), f"autoencoder_fold_{fold + 1}.pth")

# Close the TensorBoard writer
writer.close()
````

````python
# Load the full dataset (if not already loaded)
full_loader = DataLoader(train_dataset, batch_size=batch_size, shuffle=True)
test_loader = DataLoader(
    dataset=test_dataset, batch_size=batch_size, shuffle=False, num_workers=2
)

# Initialize a new model and optimizer
final_model = Autoencoder().to(device)
final_optimizer = optim.Adam(final_model.parameters(), lr=learning_rate)

# Initialize the loss function
criterion = nn.MSELoss()

# TensorBoard writer for the final model
final_writer = SummaryWriter("runs/autoencoder_final")

# Number of epochs for final training
final_num_epochs = 10  # You may increase this number

# Training loop
for epoch in range(final_num_epochs):
    train_loss = train_epoch(
        final_model, full_loader, criterion, final_optimizer, device
    )
    val_loss = validate_epoch(final_model, test_loader, criterion, device)

    # Log the losses to TensorBoard
    final_writer.add_scalars(
        "Final",
        {
            "Training": train_loss,
            "Test": val_loss,
        },
        epoch,
    )

    print(
        f"Epoch {epoch + 1}/{num_epochs}: Training Loss={train_loss:.6f}, Validation Loss={val_loss:.6f}"
    )

# Save the final model
torch.save(final_model.state_dict(), "autoencoder_final.pth")

# Close the TensorBoard writer
final_writer.close()
````

**Explanation:**

- **model = Autoencoder()**: Creates an instance of the Autoencoder class.
- **.to(device)**: Moves the model parameters to the specified device (CPU, GPU, or MPS).

````python
# Define the loss function and optimizer
criterion = nn.MSELoss()  # Mean Squared Error loss function
optimizer = torch.optim.Adam(
    model.parameters(), lr=1e-3, weight_decay=1e-5
)  # Optimizer
````

**Explanation:**

- **criterion**: The loss function used to measure the difference between the input and reconstruction. Here, Mean Squared Error (MSE) is used.
- **optimizer**: Uses the Adam optimizer to adjust the weights based on the gradients.
    - **model.parameters()**: Passes the model's parameters to the optimizer.
    - **lr=1e-3**: Sets the learning rate to 0.001.
    - **weight_decay=1e-5**: Adds L2 regularization (weight decay) to prevent overfitting.

**Explanation:**

- **transforms.Compose**: Composes several transforms together.
- **transforms.ToTensor()**: Converts PIL images or numpy arrays to PyTorch tensors and scales pixel values to [0, 1].

**Explanation:**

    - **train=True**: Specifies loading of the training set.
    - **transform=transform**: Applies the defined transformations to the data.
    - **download=True**: Downloads the dataset if it is not already present in the specified root directory.

````python
# Create a DataLoader to load the data in batches
train_loader = DataLoader(
    dataset=train_dataset,
    batch_size=128,  # Number of samples per batch
    shuffle=True,  # Shuffle the data at every epoch
)
````

**Explanation:**

- **DataLoader**: Wraps a dataset and provides an iterable over the dataset with support for batching, shuffling, etc.
    - **dataset=train_dataset**: Specifies the dataset to load.
    - **batch_size=128**: Number of samples per batch.
    - **shuffle=True**: Shuffles the data every epoch to improve randomness.

````python
# Training loop
num_epochs = 20  # Number of times to iterate over the dataset
````

**Explanation:**

- **num_epochs = 20**: Sets the number of epochs (full passes over the training data) to train the model.

````python
for epoch in range(num_epochs):
    for data in train_loader:
        img, _ = (
            data  # img contains the images, _ contains the labels (which we don't need)
        )
        img = img.view(img.size(0), -1)  # Flatten the images into vectors
        img = img.to(device)  # Move images to the device
        # ===================forward=====================
        output = model(img)  # Get the reconstructed images
        loss = criterion(output, img)  # Compute the loss between output and input
        # ===================backward====================
        optimizer.zero_grad()  # Clear the gradients
        loss.backward()  # Backpropagate the error
        optimizer.step()  # Update the weights
    # ===============log========================
    print("epoch [{}/{}], loss:{:.4f}".format(epoch + 1, num_epochs, loss.item()))
````

**Explanation:**

- **for epoch in range(num_epochs):**: Loops over each epoch.
- **for data in train_loader:**: Iterates over batches of data.
    - **data**: A batch of data, containing images and labels.
- **img, _ = data**: Unpacks the data; `img` contains images, `_` ignores labels (unused).
- **img = img.view(img.size(0), -1)**: Reshapes the images to flatten them into vectors.
    - **img.size(0)**: The batch size.
    - **-1**: Automatically calculates the size needed to flatten the image (here, 28*28=784).
- **img = img.to(device)**: Moves the images to the device (CPU or GPU).
- **output = model(img)**: Passes the images through the autoencoder model to get reconstructed outputs.
- **loss = criterion(output, img)**: Computes the MSE loss between the reconstructed images and the original images.
- **optimizer.zero_grad()**: Clears old gradients from previous steps.
- **loss.backward()**: Computes the gradient of the loss w.r.t. model parameters.
- **optimizer.step()**: Updates the model parameters based on the computed gradients.
- **print(...)**: Outputs the epoch number and current loss.

### Step 2: Obtain Latent Codes of the Selected Images

````python
# Retrieve the images using the indices
image_3, label_3 = test_dataset[digit_3_idx[0]]
image_5, label_5 = test_dataset[digit_5_idx[0]]

# Prepare the images
image_3_flat = image_3.view(1, -1).to(device)
image_5_flat = image_5.view(1, -1).to(device)

# Obtain latent codes
with torch.no_grad():
    code_3 = model.encoder(image_3_flat)
    code_5 = model.encoder(image_5_flat)
````

**Explanation:**

- **image_3, label_3 = test_dataset[digit_3_idx[0]]**: Retrieves the image and label of the first '3'.

- **image_5, label_5 = test_dataset[digit_5_idx[0]]**: Retrieves the image and label of the first '5'.

- **image_3_flat = image_3.view(1, -1).to(device)**: Flattens the image into a vector and moves it to the device.

    - **1**: Represents the batch size (since we're dealing with one image).

- **image_5_flat = image_5.view(1, -1).to(device)**: Similarly prepares the image of '5'.

- **with torch.no_grad():**: Disables gradient computation as we are only performing inference.

    - **code_3 = model.encoder(image_3_flat)**: Obtains the latent code of image '3'.

    - **code_5 = model.encoder(image_5_flat)**: Obtains the latent code of image '5'.

### Step 3: Perform Linear Interpolation in Code Space

````python
# Define the number of interpolation steps
num_interpolations = 10

# Create interpolation weights
alphas = np.linspace(0, 1, num_interpolations)

# Initialize list to store interpolated codes and reconstructed images
interpolated_codes = []
reconstructed_images = []

# Interpolate between the two codes
for alpha in alphas:
    # Compute the interpolated code
    interpolated_code = (1 - alpha) * code_3 + alpha * code_5
    interpolated_codes.append(interpolated_code)
    # Decode the interpolated code to get the image
    with torch.no_grad():
        reconstructed = model.decoder(interpolated_code)
    reconstructed_images.append(reconstructed)
````

**Explanation:**

- **num_interpolations = 10**: Sets the number of steps between the two codes.

- **alphas = np.linspace(0, 1, num_interpolations)**: Creates an array of interpolation weights ranging from 0 to 1.

    - **np.linspace(start, stop, num)**: Generates `num` evenly spaced values from `start` to `stop`.

- **interpolated_codes = []**: Initializes a list to store interpolated latent codes.

- **reconstructed_images = []**: Initializes a list to store the reconstructed images from the interpolated codes.

- **for alpha in alphas:**: Iterates over the interpolation weights.

    - **interpolated_code = (1 - alpha) * code_3 + alpha * code_5**: Computes the weighted average between `code_3` and `code_5`. When `alpha` is 0, the code is `code_3`; when `alpha` is 1, it's `code_5`.

    - **interpolated_codes.append(interpolated_code)**: Stores the interpolated code.

    - **with torch.no_grad():**: Disables gradient computation.

        - **reconstructed = model.decoder(interpolated_code)**: Decodes the interpolated code to reconstruct the image.

    - **reconstructed_images.append(reconstructed)**: Stores the reconstructed image.
