```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import multivariate_normal

# Define the means and covariances for the three Gaussian components
mean1 = [0, 0]
cov1 = [[1, 0], [0, 1]]

mean2 = [3, 3]
cov2 = [[1, 0.8], [0.8, 1]]

mean3 = [-3, 3]
cov3 = [[1, -0.8], [-0.8, 1]]

# Create a grid of (x, y) coordinates over which to evaluate the distributions
x, y = np.mgrid[-6:6:0.1, -1:8:0.1]
pos = np.dstack((x, y))

# Create multivariate normal distributions for each component
rv1 = multivariate_normal(mean1, cov1)
rv2 = multivariate_normal(mean2, cov2)
rv3 = multivariate_normal(mean3, cov3)

# Evaluate the probability density function (pdf) over the grid for each component
z1 = rv1.pdf(pos)
z2 = rv2.pdf(pos)
z3 = rv3.pdf(pos)

# Plot the contour lines for each Gaussian component
plt.figure(figsize=(10, 8))
plt.contour(x, y, z1, colors="red")
plt.contour(x, y, z2, colors="blue")
plt.contour(x, y, z3, colors="green")

# Customize the plot
plt.title("Contour Plot of a 3-Component 2D Gaussian Mixture Model")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.grid(True)
plt.show()
```
