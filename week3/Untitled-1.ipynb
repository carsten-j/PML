```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import multivariate_normal

# Parameters
weights = [0.5, 0.3, 0.2]
means = np.array([[0.2, 0.4], [0.5, 0.5], [0.7, 0.6]])
covariances = np.array(
    [
        [[0.01, 0.01], [0.01, 0.02]],
        [[0.01, -0.01], [-0.01, 0.02]],
        [[0.01, 0.01], [0.01, 0.02]],
    ]
)
colors = ["red", "green", "blue"]

# Plot
fig, ax = plt.subplots()
x, y = np.mgrid[0:1:0.01, 0:1:0.01]
pos = np.dstack((x, y))

for i in range(len(weights)):
    rv = multivariate_normal(means[i], covariances[i])
    ax.contour(x, y, rv.pdf(pos), levels=5, colors=colors[i])
    ax.text(means[i][0], means[i][1], str(weights[i]), color=colors[i], fontsize=12)

ax.set_xlim(-0.1, 1)
ax.set_ylim(0, 1)
ax.set_aspect("equal")
# ax.text(-0.1, 1.05, fontsize=12, transform=ax.transAxes)
plt.show()
```

```python
def hex_to_rgb(hex_color):
    """Convert a hex color to an RGB tuple."""
    hex_color = hex_color.lstrip("#")
    return tuple(int(hex_color[i : i + 2], 16) for i in (0, 2, 4))


def rgb_to_hex(rgb_color):
    """Convert an RGB tuple to a hex color."""
    return "#{:02x}{:02x}{:02x}".format(*rgb_color)


def weighted_color(hex1, hex2, weight1, weight2):
    """Compute the weighted color average of two hex colors."""
    rgb1 = hex_to_rgb(hex1)
    rgb2 = hex_to_rgb(hex2)

    weighted_rgb = (
        int(rgb1[0] * weight1 + rgb2[0] * weight2),
        int(rgb1[1] * weight1 + rgb2[1] * weight2),
        int(rgb1[2] * weight1 + rgb2[2] * weight2),
    )

    return rgb_to_hex(weighted_rgb)


# Example usage
hex1 = "#ff5733"  # Example color 1
hex2 = "#33b5ff"  # Example color 2
weight1 = 0.3
weight2 = 0.7
```

```python
resulting_hex = weighted_color(hex1, hex2, weight1, weight2)
print(resulting_hex)
```
