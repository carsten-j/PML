```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Define the range of x values
x = np.linspace(-10, 10, 1000)

# Compute the standard Cauchy PDF
pdf = 1 / (np.pi * (1 + x**2))

# Plot the PDF
plt.figure(figsize=(10, 6))
plt.plot(x, pdf, label="Standard Cauchy PDF", color="blue")
plt.title("Probability Density Function of the Standard Cauchy Distribution")
plt.xlabel("x")
plt.ylabel("Probability Density")
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Number of samples
n_samples = 10000  # Increased the number of samples

# Generate uniform random numbers U between 0 and 1
U = np.random.uniform(0, 1, n_samples)

# Compute the samples using the inverse CDF
samples = np.tan(np.pi * (U - 0.5))


#### **Filtering Samples for Visualization**


# Limit samples to the range [-10, 10] for visualization
sample_range = 10
samples_within_range = samples[np.abs(samples) <= sample_range]


#### **Plotting the Histogram with PDF Overlay**

# Plotting the histogram of the samples
plt.figure(figsize=(10, 6))

# Define bins for the histogram within the specified range
bins = np.linspace(-sample_range, sample_range, 200)  # Adjusted number of bins

# Plot the histogram
plt.hist(
    samples_within_range,
    bins=bins,
    density=True,
    alpha=0.6,
    color="green",
    label="Histogram of Samples",
)

# Overlay the PDF of the standard Cauchy distribution
x_plot = np.linspace(-sample_range, sample_range, 1000)
pdf_plot = 1 / (np.pi * (1 + x_plot**2))
plt.plot(x_plot, pdf_plot, "r-", linewidth=2, label="Standard Cauchy PDF")

plt.title("Histogram of Cauchy Samples with PDF Overlay")
plt.xlabel("x")
plt.ylabel("Density")
plt.legend()
plt.grid(True)
plt.xlim(-sample_range, sample_range)
plt.savefig("cauchy.png", dpi=600)  # Set x-axis limits for better visualization
plt.show()
```
